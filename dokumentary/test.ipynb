```python
import numpy as np
import cv2
from skimage import io
from matplotlib import pyplot as plt

from skimage import filters
from skimage.data import camera
from skimage.util import compare_images
```

```python
img = io.imread('https://raw.githubusercontent.com/svenschultze/mediaprocessing-notebooks/master/pictures/Lenna.png')
plt.imshow(img)
plt.show()
```

```python
edge_sobel = filters.sobel(img)
plt.imshow(edge_sobel)
plt.show()
```

```python

  
# Convert image to image gray 
tmp = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
  
# Applying thresholding technique 
_, alpha = cv2.threshold(tmp, 0, 255, cv2.THRESH_BINARY) 
  
# Using cv2.split() to split channels  
# of coloured image 
b, g, r = cv2.split(img) 
  
# Making list of Red, Green, Blue 
# Channels and alpha 
rgba = [b, g, r, alpha] 
  
# Using cv2.merge() to merge rgba 
# into a coloured/multi-channeled image 
dst = cv2.merge(rgba, 4) 
  
# Writing and saving to a new image 
plt.imshow(dst)
plt.show()
```

```python
import cv2
import cvzone
from cvzone.SelfiSegmentationModule import SelfiSegmentation

segmentor = SelfiSegmentation()

# read image
imgOffice = cv2.imread('office.jpg')

#resize office to 640x480
imgOffice = cv2.resize(imgOffice, (640, 480))

green = (0, 255, 0)

imgNoBg = segmentor.removeBG(imgOffice, green, threshold=0.50)

# show both images
cv2.imshow('office',imgOffice)
cv2.imshow('office no bg',imgNoBg)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
