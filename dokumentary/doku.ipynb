# Dokumentation Projekt "Line Art"

Gruppe: Alina Meyer, John-Uwe Riecken, Talea Schweers

## Motivation

Ein neuer Trend auf Instagram oder TikTok besteht darin die Liebsten eines Menschen in Line Art darzustellen. Das können Bilder von Tieren sowie beispielsweise Hunden oder Katzen oder auch Pärchenbilder sein. Viele Unternehmen werben damit das LineArt eines Fotos auf einem Pullover oder auch einer Kette darstellen zu können. Genau diesen Prozess wollen wir in unserem Projekt nachbilden und aus realen Fotos die dazu passende LineArts schaffen.

## Zielsetzung

Das Ziel dieses Projektes ist es, Fotos so zu manipulieren, dass diese dem Stil einer hangezeichneten Linie so nah wie möglich kommen. Dazu soll ein Programm zum Konvertieren von Fotos in ein handgezeichnetes Linienstiel Foto erstellt werden. Dabei soll ebenfalls das Gleichgewicht zwischen der Aufrechterhaltung der Essenz des Originalfotos und die Anwendung einer handgezeichneten Ästhetik beizubehalten, erforschen werden. Abschließend soll eine benutzerfreundliche Web-App für einfaches Hochladen von Fotos und Strichzeichnungskonvertierung entwickelt werden.

Darüber hinaus haben wir eigene Ziele für das Projekt festgelegt. Wir wollen verschiedene Granularitäten, also Feinheiten der LineArt und Farben implementieren. Dies soll über Einstellungsmöglichkeiten in der Oberfläche anpassbar sein.

## Meilensteine

Zur Erreichung unseres Ziel haben wir verschiedene Meilensteine aufgestellt. Unser erster Meilenstein ist die Festlegung des Ziels und des Zeitplans. Dies sollte am achten November abgeschlossen sein. In diesem Schritt haben wir die folgenden verschiedenen Meilensteine erstellt. Zudem haben wir die Ziele und Anforderungen definiert. Darüber hinaus haben wir das Projekt aufgesetzt, die ersten Operationen an den Bildern ausprobiert und die Mock-ups für das Frontend erstellt.

Der zweite Meilenstein ist die Erstellung eines Prototyps. Dieser sollte bis zum 22. November fertig sein. Dazu gehört, dass die Bilder ausgeschnitten werden. Das bedeutet der Hintergrund soll vom Frontend getrennt werden können. Als nächstes soll die Struktur des Frontends vorhanden sein soll und der Login soll auf eine neue Seite verschoben werden.

Der nächste Schritt ist die Definition von Operationen auf Bildern. In diesem Meilenstein sollen die Konturen der Bilder dargestellt werden können. Dies ist ein erster Schritt, um die LineArt darzustellen. Darüber hinaus sollen Segmente in den Bildern erkannt werden, die später beispielsweise mit verschiedenen Farben gefüllt werden können.

Im vierten Meilenstein soll das Frontend fertiggestellt werden. Dies soll noch vor Weihnachten am 20. Dezember abgeschlossen sein. Dazu gehört ein funktionierendes Frontend auf Basis der Mock-Ups. Außerdem soll es möglich sein, die Bilder mit Hilfe von KI zu bearbeiten.

Daran schließt sich der sechste Meilenstein an: die Fertigstellung der KI. Bis zum 16. Januar wollen wir in der Lage sein, die Bilder nach der Konvertierung in der Zielform anzuzeigen. Damit ist die Konvertierung eines Fotos in LineARt gemeint. Darüber hinaus wird das Frontend auf alle Funktionen getestet.

Schließlich soll der Prototyp bis zum 30. Januar 2024 fertiggestellt werden. Es sollen auch verschiedene Einstellungsmöglichkeiten für die Konvertierung implementiert werden. Dazu zählt die Einstellung verschiedener Granularitäten und verschiedener Farben. Außerdem soll es ein fertiges benutzerfreundliches Frontend geben.



## Zusammenarbeit

Des Weiteren haben wir unsere Zusammenarbeit im Projekt festgelegt. Wir wollen uns jeden Mittwoch nach der Vorlesung treffen um neue Aufgaben zu besprechen. Zum Thema Aufgabenaufteilung haben alle Teammitglieder keine feste Aufgabe, sondenr sind in allen themengebieten involviert. Die Kommunikation erfolgt über einen Gruppenchat in dem auch kurzfristige Fragen gestellt werden können. Daudurch wird ein regelmäßiger Austausch außerhalb der Treffen ermöglicht. 
Zudem werden wir paralell zum Projekt die Dokumentation anfertigen. Diese und weitere Dokumente sind über OneDrive von allen Teammitgliedern einsehbar.

Schließelich haben wir verschiedene Git Konventionen festgelegt. Wir wollen pro neue Aufgabe einen eigenen Branch machen und wenn die Aufgabe abgeschlossen ist, soll ein neuer Pull Request erstellt werden. Zudem haben wir verschiedene Konventionen zum Commiten festgelegt. Neue Features sollen mit feat abgekürzt werden und Probleme mit fix.
